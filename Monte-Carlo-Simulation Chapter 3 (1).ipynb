## 3.5.2 Construct a Monte-Carlo simulation program to calculate the area under the curve of the function y = sqrt(x) / (1 + 0.1sin(πx)) between the limits x = 0 and x = 1 (the area is 0.626, correct to three significant figures).

Method: Hit or Miss

```python
import numpy as np
from numba import njit
import time

@njit
def hitormiss(n,r,h):
  hits = np.full(n, 0)

  for i in range(n):
    rand_x = r * np.random.rand()
    rand_y = h * np.random.rand()
  
    hits[i] = rand_y <= (np.sqrt(rand_x)/(1 + 0.1 * np.sin(np.pi * rand_x)))

  a = np.count_nonzero(hits)
  return (a/n) * r * h

n = int(input('Please enter the number of trials: '))
r = float(input('Please enter range for integration: '))
h = 1

start = time.time()

print(hitormiss(n,r,h))

end = time.time()

print(f'Runtime: {end - start} seconds.')
```

## 3.5.2 Construct a Monte-Carlo simulation program to calculate the area under the curve of the function y = sqrt(x) / (1 + 0.1sin(πx)) between the limits x = 0 and x = 1 (the area is 0.626, correct to three significant figures).

Method: Sample Mean

```python
import numpy as np
from numba import njit
import time

@njit
def hitormiss2(n,r,h):

  values = []

  for i in range(n):
    rand_x = r * np.random.rand()
    

    a = (np.sqrt(rand_x)/(1 + 0.1 * np.sin(np.pi * rand_x)))
    values.append(a)
  valuesn = np.array(values)
  return (np.sum(valuesn) / n) * r

n = int(input('Please enter the number of trials: '))
r = float(input('Please enter range for integration: '))
h = 1

start = time.time()

print(hitormiss2(n,r,h))

end = time.time()

print(f'Runtime: {end - start} seconds.')
```
